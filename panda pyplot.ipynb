MODIFYING DATA

```python
# Drop irrelevant columns since people would not want to see every single thing there is
car = car.drop(["Engine Fuel Type", "Market Category", "Vehicle Style", "Popularity", "Number of Doors", "Vehicle Size"], axis = 1)
car.head(5)
```

```python
# Renaming columns (fields)
car = car.rename(columns = {

    "Engine HP": "HP",
    "Engine Cylinders": "Cylinders",
    "Transmission Type": "Transmission",
    "Driven_Wheels": "Drive Mode", 
    "highway MPG": "MPG-H",
    "city mpg": "MPG-C", 
    "MSRP": "Price",

})

car.head(5)
```

CLEANING DATA

```python
# Total number of rows and columns
car.shape
```

```python
# Rows containing duplicate data
# By using keep = first, the first occuring value is false while the remaining duplicated value is true (vice versa for last)
duplicate_row_car = car[car.duplicated(subset = None, keep = "first")]
print(f"Number of duplicate rows: {duplicate_row_car.shape}")
print(duplicate_row_car) # Shows all the duplicated rows
print(car.duplicated()) # Shows how the function works
```

```python
# Count the rows before removing the data
print(car.count())
count = 0

for i in car.count():
    print(i) # Accessing the total value for each field
    count += 1
    
    if count > 30:
        break
```

```python
# Drop (remove) the duplicates
car = car.drop_duplicates()
print(car.head(5))
print(car.count())

# Checking duplicates
duplicate_row_car = car[car.duplicated(subset = None, keep = "first")]
print(f"Number of duplicate rows: {duplicate_row_car.shape}")
car.shape
```

```python
# Find the null values
print(car.isnull().sum())
print(car.count())
```

```python
# Removing the null values
car = car.dropna()
print(car.isnull().sum())
print(car.count())
```

```python
# Plotting a Histogram for number of cars per round
car.Make.value_counts().nlargest(40).plot(kind = "bar", figsize = (10, 5))
plt.title("Number of cars by make")
plt.ylabel("Number of cars")
plt.xlabel("Make")
```

```python
# Find the relationships between the variables
plt.figure(figsize = (20, 10))
# The higher the number, the higher the correlation
# Correlation is a statistical measure that expresses the extent to which two variables are linearly related (means they change together at a constant rate)
c = car.corr() 
sns.heatmap(c, cmap = "BrBG", annot = True)
c
```
